```python
import pandas as pd

# Load the CSV file into a DataFrame
data = pd.read_csv('varydiv.csv', index_col="Task")

# Delete columns except 'Task' and those ending in 'Test Acc'
columns_to_keep = [col for col in data.columns if col.endswith('Test Acc')]

filtered_data = data[columns_to_keep]
filtered_data.head()
```

```python
import numpy as np
# Assuming you already have the filtered_data DataFrame

# Calculate the row-wise mean
row_mean = filtered_data.select_dtypes(include=[np.number]).mean(axis=0)

# Convert the row-wise mean into a Series
series_data = pd.Series(row_mean, index=filtered_data.columns)
series_data = series_data.rename(index=lambda x: x.replace("VaryDiv-", "").replace(" - Test Acc", ""))
series_data = series_data.sort_index()
series_data = series_data*100

# Set the display options for pandas
pd.set_option('display.float_format', '{:.5f}'.format)
# Print the nicely formatted Series
print(len(series_data))
print(series_data)
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
data1 = pd.read_csv('varydivForget.csv', index_col="Task")

# Delete columns except 'Task' and those ending in 'Test Acc'
columns_to_keep1 = [col for col in data1.columns if col.endswith('Test Acc')]
filtered_data1 = data1[columns_to_keep1]

for column in filtered_data1.columns:
    first_value = filtered_data1[column].iloc[0]
    filtered_data1.loc[:, column] = (first_value - filtered_data1[column]) / first_value

filtered_data1.head()
```
