```python
import pickle
import os 
import matplotlib.pyplot as plt
import numpy as np

path = "./Results/VaryBufferSelection"
dirlist = os.listdir(path)

dirlist.sort()
```

# classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck') 

```python
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
import statistics
fic = {}
def sum_lists(*args):
    return list(map(sum, zip(*args)))

# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
dirwithoutruns = set([x.split("-", 1)[0] for x in dirlist])
dirwithoutruns = list(dirwithoutruns)
nonshuffledirwithoutruns = sorted([n for n in dirwithoutruns if "shuffle" not in n])
shuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" in n]
print(nonshuffledirwithoutruns)
labels = ['Distance Vector-Easiest', 'Distance Vector-Hardest', 'Distance Vector-Uniform', 'Confidence Score-Easiest', 'Confidence Score-Hardest', 'Confidence Score-Uniform']
labelsforlater = ['Distance \n Vector \n Easiest', 'Distance \n Vector \n Hardest', 'Distance \n Vector \n Uniform', 'Confidence \n Score \n Easiest', 'Confidence \n Score \n Hardest', 'Confidence \n Score \n Uniform']


for dir in nonshuffledirwithoutruns:
  lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,7,70,700,7000,70000]
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["plane"]
    yaa = [yaa[ii*250+249] for ii in [0,1,2,3,4]]
    ybb=aa["car"]
    ybb = [ybb[ii*250+249] for ii in [0,1,2,3,4]]

    yavg = sum_lists(yaa, ybb)
    # print(yaa, ybb, yavg)
    yavg = [ff/2 for ff in yavg]
    yavg = [(yavg[0]-x)/yavg[0]*100 for x in yavg]
    lisofacc.append(yavg)

  accperrun = []
  # print(dir)
  for i in lisofacc:
    # print(i)
    accperrun.append(sum(i))
  accperrun = [ii/4 for ii in accperrun]
  # print(accperrun)
  fic[dir] = (statistics.mean(accperrun), statistics.stdev(accperrun))
print(fic)
```

```python
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
def sum_lists(*args):
    return list(map(sum, zip(*args)))
    
fig, ax = plt.subplots(figsize=(14,11))
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
dirwithoutruns = set([x.split("-", 1)[0] for x in dirlist])

dirwithoutruns = list(dirwithoutruns)
nonshuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" not in n]
nonshuffledirwithoutruns.sort()
shuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" in n]
shuffledirwithoutruns.sort()

print(nonshuffledirwithoutruns)

labels = ['Distance Vector-Easiest', 'Distance Vector-Hardest', 'Distance Vector-Uniform', 'Confidence Score-Easiest', 'Confidence Score-Hardest', 'Confidence Score-Uniform']
labelsforlater = ['Distance \n Vector \nEasiest', 'Distance \n Vector \nHardest', 'Distance \n Vector \nUniform', 'Confidence \n Score \nEasiest', 'Confidence \n Score \nHardest', 'Confidence \n Score \nUniform']

x = [1,2,3,4,5]
for dir in nonshuffledirwithoutruns:
  lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,7,70,700,7000,70000]
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["plane"]
    lisofacc.append([yaa[ii*250+249] for ii in [0,1,2,3,4]])

  yavg = [0,0,0,0,0]
  for i in lisofacc:
    yavg = sum_lists(yavg, i)
  
  yavg = [x/20 for x in yavg]
  yavgf = [(yavg[0]-x)/yavg[0]*100 for x in yavg]

  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["car"]
    lisofacc.append([yaa[ii*250+249] for ii in [0,1,2,3,4]])
    
  yavg = [0,0,0,0,0]
  for i in lisofacc:
    yavg = sum_lists(yavg, i)
  
  yavg = [x/20 for x in yavg]
  yavgf2 = [(yavg[0]-x)/yavg[0]*100 for x in yavg]
  
  def sum_lists(*args):
    return list(map(sum, zip(*args)))
  ax.plot(x, list(y/2 for y in sum_lists(yavgf, yavgf2)), label=labels[nonshuffledirwithoutruns.index(dir)])
  
  


lisofshuffle = []
lisofshufflec = []
for dir in shuffledirwithoutruns:
  lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,7,70,700,7000,70000]
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["plane"]
    lisofacc.append([yaa[ii*250+249] for ii in [0,1,2,3,4]])

  yavg = [0,0,0,0,0]
  for i in lisofacc:
    yavg = sum_lists(yavg, i)
  
  yavg = [x/20 for x in yavg]
  lisofshuffle.append(yavg)

  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["car"]
    lisofacc.append([yaa[ii*250+249] for ii in [0,1,2,3,4]])
    
  yavg = [0,0,0,0,0]
  for i in lisofacc:
    yavg = sum_lists(yavg, i)
  
  yavg = [x/20 for x in yavg]
  lisofshufflec.append(yavg)

import numpy
lisofshuffle = numpy.array(lisofshuffle)
lisofshuffle = lisofshuffle.transpose()
import statistics
plan = [statistics.mean(x) for x in lisofshuffle]
plane = [(plan[0]-x)/plan[0]*100 for x in plan]
lisofshufflec = numpy.array(lisofshufflec)
lisofshufflec = lisofshufflec.transpose()
ca = [statistics.mean(x) for x in lisofshufflec]
car = [(ca[0]-x)/ca[0]*100 for x in ca]
ax.plot(x, list(y/2 for y in sum_lists(plane, car)), label="Baseline", color="k")



# ax.set_title("Percentage of Forgetting (%) against Task", fontsize=20)
ax.legend(fontsize=20)
ax.set_xticks(np.arange(min(x), max(x)+1), fontsize=20)
ax.grid()
ax.set_xticklabels([1,2,3,4,5],fontsize=20)
ax.set_yticklabels([0,20,40,60,80,100], fontsize=20)
ax.set_xlabel("Task", fontsize=20)
ax.set_ylabel("Percentage of Forgetting (%)", fontsize=20) #Avg Accuracy up to current task
ax.set_ylim(top=100, bottom=0) # Font size

fig.savefig("./Percentage of Forgetting (%) against Task at EndPoint.svg", format = 'svg')
```
