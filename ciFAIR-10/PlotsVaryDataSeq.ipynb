```python
import pickle
import os 
import matplotlib.pyplot as plt
import numpy as np
    
path = "./Results/VaryReplayClassSeq"
dirlist = os.listdir(path)

dirlist.sort()
# dirlist = [x for x in dirlist if "-seed7" not in x and "-seed3" not in x]
```

```python
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot (Dict data)

import statistics 
def sum_lists(*args):
    return list(map(sum, zip(*args)))
    
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
dirwithoutruns = set([x.split("-", 1)[0] for x in dirlist])

dirwithoutruns = list(dirwithoutruns)
nonshuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" not in n]
nonshuffledirwithoutruns.sort()
shuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" in n]
shuffledirwithoutruns.sort()


labels = ['Instance-Distance Vector-ez2hard', 'Instance-Distance Vector-hard2ez', 'Class-Distance Vector-ez2hard', 'Class-Distance Vector-hard2ez','Instance-Confidence Score-ez2hard', 'Instance-Confidence Score-hard2ez', 'Class-Confidence Score-ez2hard', 'Class-Confidence Score-hard2ez']
labelsforlater = ['Instance \n Distance Vector \n ez2hard', 'Instance \n Distance Vector \n hard2ez', 'Class \n Distance Vector \n ez2hard', 'Class \n Distance Vector \n hard2ez','Instance \n Confidence Score \n ez2hard',   'Instance \n Confidence Score \n hard2ez', 'Class \n Confidence Score \n ez2hard', 'Class \n Confidence Score \n hard2ez']

x = [1,2,3,4,5]
fic = {}
for dir in nonshuffledirwithoutruns:
  lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,7,70,700,7000,70000]
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["plane"]
    yaa = [yaa[ii*250+249] for ii in [0,1,2,3,4]]
    ybb=aa["car"]
    ybb = [ybb[ii*250+249] for ii in [0,1,2,3,4]]

    yavg = sum_lists(yaa, ybb)
    # print(yaa, ybb, yavg)
    yavg = [ff/2 for ff in yavg]
    yavg = [(yavg[0]-x)/yavg[0]*100 for x in yavg]
    lisofacc.append(yavg)

  accperrun = []
  # print(dir)
  for i in lisofacc:
    # print(i)
    accperrun.append(sum(i))
  accperrun = [ii/4 for ii in accperrun]
  # print(accperrun)
  fic[dir] = (statistics.mean(accperrun), statistics.stdev(accperrun))
print(fic)

```
