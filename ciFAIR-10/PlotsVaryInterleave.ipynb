```python
import pickle
import os 
import matplotlib.pyplot as plt
import numpy as np

path = "./Results/VaryInterleave"
dirlist = os.listdir(path)

dirlist.sort()
dirlist = [x for x in dirlist]
print(dirlist)
# print(dirlist)
```

```python
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
import statistics

fic = {}
def sum_lists(*args):
    return list(map(sum, zip(*args)))
    
x = [1, 2, 3, 4, 5]
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
dirwithoutruns = set([x.split("-", 1)[0] for x in dirlist])
dirwithoutruns = list(dirwithoutruns)
nonshuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" not in n]
nonshuffledirwithoutruns.sort(key=lambda x:int(x[6:]))
shuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" in n]
shuffledirwithoutruns.sort(key=lambda x:int(x[11:]))

x = [1,2,3,4,5]
for dir in nonshuffledirwithoutruns:
  lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,8,80,800,8000,80000]
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Accuracy by class.pkl', 'rb'))
    yaa=aa["plane"]
    yaa = [yaa[ii*250+249] for ii in [0,1,2,3,4]]
    ybb=aa["car"]
    ybb = [ybb[ii*250+249] for ii in [0,1,2,3,4]]

    yavg = sum_lists(yaa, ybb)
    # print(yaa, ybb, yavg)
    yavg = [ff/2 for ff in yavg]
    yavg = [(yavg[0]-x)/yavg[0]*100 for x in yavg]
    lisofacc.append(yavg)

  accperrun = []
  # print(dir)
  for i in lisofacc:
    # print(i)
    accperrun.append(sum(i))
  accperrun = [ii/4 for ii in accperrun]
  # print(accperrun)
  fic[dir] = statistics.mean(accperrun)
print(fic)
```

```python
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot

x = [1, 2, 3, 4, 5]


fig, ax = plt.subplots(figsize=(14,11))
# Average of Total Accuracy Plots accross each replay sequence --> 1 Plot
dirwithoutruns = set([x.split("-", 1)[0] for x in dirlist])
dirwithoutruns = list(dirwithoutruns)
nonshuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" not in n]
nonshuffledirwithoutruns.sort(key=lambda x:int(x[6:]))
shuffledirwithoutruns = [n for n in dirwithoutruns if "shuffle" in n]
shuffledirwithoutruns.sort(key=lambda x:int(x[11:]))

print(nonshuffledirwithoutruns)

valueforanotherplot5 = {}
valueforanotherplot4 = {}
valueforanotherplot3 = {}
valueforanotherplot2 = {}

valueforanotherplot5s = {}
valueforanotherplot4s = {}
valueforanotherplot3s = {}
valueforanotherplot2s = {}

colorlis = ["#000000","#330000","#660000","#990000","#cc0000","#ff0000","#ff3333","#ff6666", "#ff9999"]
colorlis.reverse()
count = 0

for dir in nonshuffledirwithoutruns:
  lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,8,80,800,8000,80000]
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Total Accuracy.pkl', 'rb'))
    yaa=aa["Accuracy"]
    lisofacc.append([yaa[ii*250+249] for ii in [0,1,2,3,4]])

  yavg = [0,0,0,0,0]
  for i in lisofacc:
    if dir in valueforanotherplot5s:
      valueforanotherplot5s[dir].append(i[-1])
    else:
      valueforanotherplot5s[dir]=[i[-1]]
    if dir in valueforanotherplot4s:
      valueforanotherplot4s[dir].append(i[-2])
    else:
      valueforanotherplot4s[dir]=[i[-2]]
    if dir in valueforanotherplot3s:
      valueforanotherplot3s[dir].append(i[-3])
    else:
      valueforanotherplot3s[dir]=[i[-3]]
    if dir in valueforanotherplot2s:
      valueforanotherplot2s[dir].append(i[-4])
    else:
      valueforanotherplot2s[dir]=[i[-4]]
    yavg = sum_lists(yavg, i)


  yavg = [x/20 for x in yavg]
  valueforanotherplot5[dir] = yavg[-1]
  valueforanotherplot4[dir] = yavg[-2]
  valueforanotherplot3[dir] = yavg[-3]
  valueforanotherplot2[dir] = yavg[-4]
  
  ax.plot(x, yavg, label=f"Division {dir[6:]}", color=colorlis[count])
  count += 1
  
valueforanotherplot5 = dict(valueforanotherplot5.items())
valueforanotherplot4 = dict(valueforanotherplot4.items()) 
valueforanotherplot3 = dict(valueforanotherplot3.items()) 
valueforanotherplot2 = dict(valueforanotherplot2.items()) 

lisofseed = [1,10,100,1000,10000,3,30,300,3000,30000,5,50,500,5000,50000,8,80,800,8000,80000]


lisofshuffle = []
for dir in shuffledirwithoutruns:
  
  lisofacc = []
  for i in lisofseed:
    aa=pickle.load(open(f'{path}/{dir}-seed{i}/Total Accuracy.pkl', 'rb'))
    yaa=aa["Accuracy"]
    lisofacc.append([yaa[ii*250+249] for ii in [0,1,2,3,4]])

  yavg = [0,0,0,0,0]
  for i in lisofacc:
    yavg = sum_lists(yavg, i)
  
  yavg = [x/20 for x in yavg]
  lisofshuffle.append(yavg)
  
  valueforanotherplot5[dir] = yavg[-1]
  valueforanotherplot4[dir] = yavg[-2]
  valueforanotherplot3[dir] = yavg[-3]
  valueforanotherplot2[dir] = yavg[-4]


import numpy
lisofshuffle = numpy.array(lisofshuffle)
lisofshuffle = lisofshuffle.transpose()
import statistics
mean = [statistics.mean(x) for x in lisofshuffle]
std = [statistics.stdev(x) for x in lisofshuffle]
ax.plot(x, mean, label="Baseline", color="c", linewidth=1)

coun = 0
for i in std:
  ax.errorbar(coun+1,mean[coun], yerr=[i],color="k", linewidth=0.8)
  coun+=1
  
ychance = [1/(i*2)*100 for i in range(1, 6)]
ax.plot(x, ychance, label="Chance Level", linestyle='dashed',color="g")

# ax.set_title("Total Accuracy (%) against Task", fontsize=20)
ax.legend(fontsize=20)
ax.set_xticks(np.arange(min(x), max(x)+1), fontsize=20)
ax.grid()
ax.set_xticklabels([1,2,3,4,5],fontsize=20)
ax.set_yticklabels([0,20,40,60,80,100], fontsize=20)
ax.set_xlabel("Task", fontsize=20)
ax.set_ylabel("Total Accuracy (%)", fontsize=20) #Avg Accuracy up to current task
ax.set_ylim(top=100, bottom=0) # Font size

fig.savefig("./Total Accuracy (%) against Task at EndPoint.svg", format = 'svg')
```
